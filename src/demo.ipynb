```python
import pandas as pd

# ==========================
# 1. Cargar los datos
# ==========================
df_mat = pd.read_csv("ucimlrepo/student-mat.csv", sep=";")
df_por = pd.read_csv("ucimlrepo/student-por.csv", sep=";")

# 2. Unir ambos datasets
df = pd.concat([df_mat, df_por], ignore_index=True)
# Mostrar las primeras 50 filas
print(df.head(50))

print("Filas totales:", len(df))
print("Columnas disponibles:", list(df.columns))
print(df.head())

# ==========================
# 3. Funciones para tablas de frecuencia
# ==========================
def tabla_frecuencia_categorica(columna):
    """
    Retorna una tabla con frecuencias absolutas y relativas para una variable categórica.
    """
    tabla = pd.DataFrame({
        "Frecuencia": df[columna].value_counts().sort_index(),
        "Porcentaje": df[columna].value_counts(normalize=True).sort_index() * 100
    })
    return tabla

def tabla_frecuencia_numerica(columna, bins):
    """
    Retorna una tabla con frecuencias absolutas y relativas para una variable numérica agrupada en intervalos.
    """
    categorias = pd.cut(df[columna], bins=bins)
    tabla = pd.DataFrame({
        "Frecuencia": categorias.value_counts().sort_index(),
        "Porcentaje": categorias.value_counts(normalize=True).sort_index() * 100
    })
    return tabla

# ==========================
# 4. Ejemplos de uso
# ==========================

# Horas de estudio (studytime: categórica)
tabla_studytime = tabla_frecuencia_categorica("studytime")
print("\n📊 Distribución de Horas de Estudio:")
print(tabla_studytime)
#Mostrar los primeros 50 valores de la variable studytime
print("\n🔎 Primeros 50 datos de studytime:")
print(df["studytime"].head(50).to_frame())

# Nota final (G3: numérica → intervalos de 0 a 20)
tabla_G3 = tabla_frecuencia_numerica("G3", bins=[0,5,10,15,20])
print("\n📊 Distribución de Notas Finales (G3):")
print(tabla_G3)
# Mostrar los primeros 50 valores de la variable G3 (nota final)
print("\n🔎 Primeros 50 datos de G3:")
print(df["G3"].head(50).to_frame())
```
